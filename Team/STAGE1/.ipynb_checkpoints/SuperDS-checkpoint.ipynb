Description:
    In this task we will read all the covid-19 datasets and reshape them to extract  data easily and present them in a user readable way. We have merged all the three covid datasets into a Super covid dataset. We have used pandas and numpy packages to to complete this task successfully.
    
   We have to make a big note of the functions we have used in this task.
        1) pd.melt() function which we have used to convert a wide dataframe into a long dataframe so that we can extract data              according to the date we want.We have referred the pd.melt() function from the pandas documentation at this location            "https://pandas.pydata.org/docs/reference/api/pandas.melt.html".
        2) pd.read_csv() function which is used to read the csv file.
        3) pd.merge() function is used to merge two dataframes based on the common column.
           syntax: pd.merge(file1, file2, on=(common columns), how='type of join')
        4) pd.to_csv() function is used to convert the dataframe into a csv file.
        

```python
# Importing pandas and numpy packages.

import pandas as pd
import numpy as np
```

```python
# Reading the covid deaths csv file using Pd.read_csv() function.

deaths= pd.read_csv("../../DATASETS/COVID DATASETS/covid_deaths_usafacts.csv")
deaths.head() #head() function displays the first 5 rows of the dataframe.
```

```python
# Reading the covid Confirmed cases csv file using pd.read_csv() function.

confirmed =  pd.read_csv("../../DATASETS/COVID DATASETS/covid_confirmed_usafacts.csv")
confirmed.head() # Displaying the first 5 rows of the covid confirmed cases dataframe.
```

```python
# Reading the Population of each county in USA.

population = pd.read_csv("../../DATASETS/COVID DATASETS/covid_county_population_usafacts.csv")
population.head() # Displaying the first 5 rows of the Population dataframe according to the countyFIPS.
```

```python
# Using pd.melt() function to convert the covid death cases wide data frame into long dataframe according to the date.
# Syntax of pd.melt() function is provided below
# longDataFrame = pd.melt(frame = wideDataFrame, id_vars=(tuple or list with columns that should not change), value_vars=(columns to be changed), var_name=(Name for the new column), value_vars=(Name for the values of the new column))
# We did not use value_vars in our code because we cannot specify hundreds of date coumns in a single.
# If we do not specify the value_vars() it takes all columns except the columns specified in id_vars().


deaths_New = pd.melt(frame = deaths, id_vars=('countyFIPS','County Name','State','StateFIPS'),var_name=["Date"],value_name='Number of Deaths')
deaths_New # This is the new covis deaths dataframe as a long dataframe.
```

```python
# Using pd.melt() function to convert the covid confirmed cases wide data frame into long dataframe according to the date.
# Syntax of pd.melt() function is provided below
# longDataFrame = pd.melt(frame = wideDataFrame, id_vars=(tuple or list with columns that should not change), value_vars=(columns to be changed), var_name=(Name for the new column), value_vars=(Name for the values of the new column))
# We did not use value_vars in our code because we cannot specify hundreds of date coumns in a single.
# If we do not specify the value_vars() it takes all columns except the columns specified in id_vars().

confirmed_New = pd.melt(frame = confirmed, id_vars=('countyFIPS','County Name','State','StateFIPS'),var_name=["Date"],value_name='Number of Cases')
confirmed_New # This is the new confirmed cases dataframe as a long dataframe.
```

```python
# In this step we have merged the confirmed_New and deaths_New dataframes on 'countyFIPS', 'County Name', 'State', 'StateFIPS', 'Date'.
# The newly merged dataframe is in 'superCovidDS' dataframe.

superCovidDS = pd.merge(confirmed_New,deaths_New, on=('countyFIPS','County Name','State','StateFIPS','Date'))
```

```python
# Here we have displayed the superCovidDS dataframe.

superCovidDS
```

```python
# From the population dataframe i have droped 'County Name', since we had merge issues with superCovidDS dataframe
# The new population dataframe is taken in 'population_New' dataframe.

population_New=population.drop(['County Name'],1)
population_New
```

```python
# Here  we have merged the 'superCovidDS' and 'population_New dataframes' through inner join.

supercovidDS = pd.merge(superCovidDS,population_New, how='inner')
```

```python
superCovidDS
```

```python
# In this step are coverting it into a csv file name 'covidSuperDS.csv'

superCovidDS.to_csv('../../Team/STAGE1/covidSuperDS.csv', index=False)
```
