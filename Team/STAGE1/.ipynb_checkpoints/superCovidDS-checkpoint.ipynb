Description: In this task we will read all the covid-19 datasets to extract data easily and present them in a user readable way. We have merged all the three covid datasets into a Super covid dataset. We have used pandas and numpy packages to to complete this task successfully.

We have to make a big note of the functions we have used in this task. 
1) pd.read_csv() function which is used to read the csv file. 
2) pd.merge() function is used to merge two dataframes based on the common column. syntax: pd.merge(file1, file2, on=(common columns), how='type of join') 
3) pd.to_csv() function is used to convert the dataframe into a csv file.

```python
# Importing numpy and pandas packages.

import numpy as np
import pandas as pd
```

```python
# Reading covid cases confirmed csv file and displaying it.

cases = pd.read_csv("../../DATASETS/COVID DATASETS/covid_confirmed_usafacts.csv")
cases
```

```python
# Reading covid deaths csv file and displaying it.

deaths = pd.read_csv("../../DATASETS/COVID DATASETS/covid_deaths_usafacts.csv")
deaths
```

```python
# reading population of each county csv file and displaying it.

population = pd.read_csv("../../DATASETS/COVID DATASETS/covid_county_population_usafacts.csv")
population
```

```python
# Removing the 'County Name', 'State' in population data to avoid merge errors.

population_New=population.drop(['County Name','State'],axis=1)
```

```python
population_New # New population data.
```

```python
# Merging cases and deaths dataframes on 'countyFIPS', 'County Name', 'State', 'StateFIPS' columns through inner join.

superCovidDS = pd.merge(cases,deaths,on=('countyFIPS','County Name','State','StateFIPS'),how='inner')
superCovidDS 

# This is the dataframe after merging.
# Here Date_X refers to cases and Date_y refers to Deaths.
```

```python
# Merging 'superCovidDS' and 'Population_New' dataframes on 'countyFIPS' column.

superCovidDS = pd.merge(superCovidDS,population_New, on=('countyFIPS'))
```

```python
superCovidDS 

# This is the dataframe after merging all the three dataframes.
```

```python
# Filtering the dataframe by removing unwanted rows.

superCovidDS = superCovidDS[superCovidDS['County Name']!='Statewide Unallocated']
```

```python
superCovidDS
```

```python
# resetting the index values.

superCovidDS=superCovidDS.reset_index()
```

```python
# Removing the unwanted 'index' column

superCovidDS.drop('index',axis=1,inplace=True)
```

```python
# Converting the dataframe into csv file and storing it.

superCovidDS.to_csv("../../Team/STAGE1/superCovidDS.csv",index=False)
```
