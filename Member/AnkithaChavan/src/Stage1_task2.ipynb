## STAGE 1 TASk 2 - TRENDS OF CASES AND DEATHS IN ALABAMA STATE

#### I have considered the Albama State and plotted the graphs for number of cases and deaths for one week period.

```python
import pandas as pd 
covid_data = pd.read_csv(r"../data/super_covid.csv", encoding= 'unicode_escape')
```

```python
#analysis of data from start of 2021 january to 2021 december end.
covid_data = covid_data[(covid_data["Date"] >= '2021-01-01') & (covid_data["Date"] < '2022-01-01')]
#Sorting the vales and resetting the index from 0
covid_data_2021= covid_data.sort_values(by=['Date']).reset_index(drop=True)
covid_data_2021
```

```python
covid_data
#Removing Index as we have multiple columns
del covid_data[covid_data.columns[0]]
#resetting index from 0 and sorting on date.
covid_data= covid_data.sort_values(by=['Date']).reset_index(drop=True)
covid_data
```

```python
#Analysis for albama state so only choosing the rows for state albama
covid_data_2021 = covid_data_2021[covid_data_2021["State"] == "AL"]
#Resetting the index to true and sorting by date
covid_data_2021_alabama= covid_data_2021.sort_values(by=['Date']).reset_index(drop=True)

covid_data_2021_alabama
```

```python
# sum of Number of cases and death in albama state grouped by date.
covid_data_2021_Alabama_cases = covid_data_2021_alabama.groupby('Date',)['Number of Cases','Number of Deaths'].sum()
covid_data_2021_Alabama_cases
```

```python
# last week cases in albama state
covid_data_2021_Alabama_cases_last_week = covid_data_2021_Alabama_cases.tail(7)
covid_data_2021_Alabama_cases_last_week
```

```python
#plotting the last weeks no of cases and death report.
covid_data_2021_Alabama_cases_last_week.plot()
```

### The above data is of Albama state from 25-12-2021 to 31-12-2021 which indicates the number of cases and deaths in Alabama state where the number of cases are increasing and the number of deaths are slightly increasing as shown in the plot and table above.
