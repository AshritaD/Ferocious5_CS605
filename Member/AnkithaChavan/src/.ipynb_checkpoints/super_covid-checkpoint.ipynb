#### As  we are merging the hospital enrichment data with covid data based on county_fips and dates columns.So, here we are using melt function to transpose the date column so that we can merge with the enrichment data.

```python
import pandas as pd
import numpy as np
```

```python
# reading the death data
deaths = pd.read_csv("../data/covid_deaths_usafacts.csv")
deaths.head()
```

```python
# reading the confirmed data
cases = pd.read_csv("../data/covid_confirmed_usafacts.csv")
cases.head()
```

```python
#reading the population data
population = pd.read_csv("../data/covid_county_population_usafacts.csv")
population.head()
```

```python
# using the melt function so that we get the all the dates in one column and merging will be easy with enrichment data.
deaths_transpose = pd.melt(frame= deaths, id_vars=('countyFIPS','County Name','State','StateFIPS'), var_name=["Date"], value_name='Number of Deaths')
deaths_transpose = deaths_transpose[deaths_transpose['countyFIPS'] != 0]
deaths_transpose
```

```python
# using the melt function so that we get the all the dates in one column and merging will be easy with enrichment data.
cases_transpose = pd.melt(frame = cases, id_vars=('countyFIPS','County Name','State','StateFIPS'),var_name=["Date"],value_name='Number of Cases')
cases_transpose = cases_transpose[cases_transpose['countyFIPS'] != 0]
cases_transpose
```

```python
#merging the cases and death data.
cases_deaths_merge = pd.merge(cases_transpose,deaths_transpose, on=('countyFIPS','County Name','State','StateFIPS','Date'))
cases_deaths_merge
```

```python
#merging the above data with poplation data.
population = population.drop(columns = ['County Name'])
super_covid = pd.merge(cases_deaths_merge,population, on=('countyFIPS', 'State'))
super_covid
```

```python
super_covid.to_csv('../data/super_covid.csv')
```
