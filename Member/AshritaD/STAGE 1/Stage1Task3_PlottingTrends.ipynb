```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#reading the data for the confirmed number of COVID cases and displaying them
confirmedCases = pd.read_csv(r"..\..\DATASETS\COVID DATASETS\covid_confirmed_usafacts.csv")
confirmedCases
```

```python
# reading the confirmed number of COVID Deaths and displaying them
confirmedDeaths = pd.read_csv(r"..\..\DATASETS\COVID DATASETS\covid_deaths_usafacts.csv")
confirmedCases
```

```python
#Extracting the last week of data from the given set(number of confirmed cases)
confirmedCasesNJ = confirmedCases[confirmedCases["State"] == "NJ"].iloc[:,-7:]
confirmedCasesNJ
```

```python
#Extracting the last week of data from the given set(number of deaths)
confirmedDeathsNJ = confirmedDeaths[confirmedDeaths["State"] == "NJ"].iloc[:,-7:]
confirmedDeathsNJ
```

```python
#Plotting the graph for to check the trend of the number of cases for the given time frame
plotCNJ = confirmedCasesNJ.sum().plot()
plotCNJ.set_title("Confirmed COVID-19 cases for the past week in NJ")
```

```python
#Plotting the graph for to check the trend of the number of deaths for the given time frame
plotCNJ = confirmedDeathsNJ.sum().plot()
plotCNJ.set_title("Confirmed COVID-19 deaths for the past week in NJ")
```
