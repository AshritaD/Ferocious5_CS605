```python
# reading two csv files
data1 = pd.read_csv(r"..\..\DATASETS\SUPER DATASETS\covidSuperDS.csv", low_memory=False)
data2= pd.read_csv(r"..\..\DATASETS\ENRICHMENT DATASETS\Enrichment_Housing1Y2021.csv", low_memory=False)

#getting the last 5 digits to match the countyfips in covidsuperdata
data2['countyFIPS'] = data2['GEO_ID'].str[-5:]
data2['countyFIPS'] = pd.to_numeric(data2['countyFIPS'])

# using merge function by setting 
output1 = pd.merge(data1, data2,
                   on= ('countyFIPS')
                   )
                   
#displaying result

output1.drop(output1.columns[output1.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
output1 = output1.drop('NAME', axis=1)

output1
```

```python
output1.to_csv(r"..\..\Member\AshritaD\EnrichmentSuperDataSet_Housing.csv", index = False)
```
