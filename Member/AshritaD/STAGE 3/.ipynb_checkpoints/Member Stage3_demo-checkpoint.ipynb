```python
from scipy.stats.distributions import gamma

plt.hist(ccnjSum,density=True, bins=9, ec='black')#, figsize=(12,6))
label = 'alpha = {0:.2f}\nbeta = {1:.2f}'.format(alpha_mom[0], beta_mom[0])
plt.annotate(label, xy=(2000, 0.0002))
plt.plot(np.linspace(0, ccnjSum.max()), gamma.pdf(np.linspace(0, ccnjSum.max()), alpha_mom[0],scale =beta_mom[0]))
```

```python
#MLE Normal distribution
n_bins = 10
plt.hist(ccnjSum, bins=n_bins, ec='black', density=True)
alpha, beta = norm.fit(ccnjSum)
label = 'alpha = {0:.2f}\nbeta = {1:.2f}'.format(alpha, beta)
plt.annotate(label, xy=(20000, 0.0002))
x = np.linspace(0, ccnjSum.max(), n_bins)
plt.plot(x, norm.pdf(x, alpha, beta))
plt.show()
```

```python
'''It is not necessarily a problem if the density line does not peak all the way up to the top of the highest histogram bar. The purpose of the density line is to provide a smooth estimate of the underlying distribution of the data, and it is possible that the data may not perfectly match any specific distribution. As long as the overall shape of the density line is consistent with what you would expect based on the data, and it captures the main features of the distribution, then it is generally acceptable. However, it's always a good idea to carefully examine your data and consider if there might be any issues or limitations with your analysis.'''
```

```python
#mom for normal distribution
n_bins = 10
plt.hist(ccnjSum, bins=n_bins, ec='black', density=True)
mu_mom = np.mean(ccnjSum)
sigma_mom = np.std(ccnjSum)
label = 'mu = {0:.2f}\nsigma = {1:.2f}'.format(mu_mom, sigma_mom)
plt.annotate(label, xy=(20000, 0.0002))
x = np.linspace(0, ccnjSum.max(), n_bins)
plt.plot(x, norm.pdf(x, mu_mom, sigma_mom))
plt.show()
```

```python
n_bins = 10
plt.hist(ccnjSum, bins=n_bins, ec='black', density=True)

# Calculate the mean and standard deviation of the data
mu = np.mean(ccnjSum)
sigma = np.std(ccnjSum)

# Plot the MoM normal distribution
x = np.linspace(0, ccnjSum.max(), n_bins)
plt.plot(x, norm.pdf(x, mu, sigma), label='MoM Normal')

# Add a label to the plot
label = 'mu = {0:.2f}\nsigma = {1:.2f}'.format(mu, sigma)
plt.annotate(label, xy=(20000, 0.0002))

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```

```python
from scipy.stats.distributions import poisson

# Calculate the mean of the data set
lmbda = np.mean(ccnjSum)

# Plot the histogram
plt.hist(ccnjSum, density=True, bins=9, ec='black') 

# Plot the Poisson distribution
x = np.arange(0, ccnjSum.max())
plt.plot(x, poisson.pmf(x, lmbda), label='Poisson')

# Add a label to the plot
label = 'lambda = {0:.2f}'.format(lmbda)
plt.annotate(label, xy=(2000, 0.0002))

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```

```python
# fit gamma distribution using MLE
alpha_mle, loc_mle, beta_mle = gamma.fit(ccnjSum, floc=0)

# generate histogram of data
n_bins = 10
hist, bins, _ = plt.hist(ccnjSum, bins=n_bins, ec='black', density=True)

# plot fitted gamma distribution
x = np.linspace(0, ccnjSum.max(), 100)
pdf = gamma.pdf(x, alpha_mle, loc_mle, beta_mle)
plt.plot(x, pdf, 'r-', linewidth=2)

# show plot
plt.show()

# print estimated parameters
print('alpha =', alpha_mle)
print('beta =', beta_mle)
```

```python
from scipy.stats import poisson
from scipy.optimize import curve_fit

# define the Poisson PDF
def poisson_pdf(x, mu):
    return poisson.pmf(x, mu)

# fit the Poisson distribution to your data using curve_fit
params, _ = curve_fit(poisson_pdf, np.arange(len(ccnjSum)), ccnjSum)

# generate histogram of data
n_bins = 10
plt.hist(ccnjSum, bins=n_bins, ec='black', density=True)

# plot fitted Poisson distribution
x = np.arange(0, len(ccnjSum))
pdf = poisson_pdf(x, *params)
plt.plot(x, pdf, 'r-', linewidth=2)

# show plot
plt.show()

# print estimated parameter
print('mu =', params[0])
```

```python
from scipy.stats import binom

n_bins = 10
plt.hist(ccnjSum, bins=n_bins, ec='black', density=True)
n = len(ccnjSum)
p_mle = ccnjSum.mean() / ccnjSum.max()  # maximum likelihood estimate of p
label = 'n = {0}\np = {1:.2f}'.format(n, p_mle)
plt.annotate(label, xy=(20000, 0.0002))
x = np.linspace(0, ccnjSum.max(), n_bins)
plt.plot(x, binom.pmf(x, n, p_mle))
plt.show()

# print estimated parameter
print('p =', p_mle)
```

```python
from scipy.stats import binom

n_bins = 10
plt.hist(ccnjSum, bins=n_bins, ec='black', density=True)

# fit binomial distribution using method of moments
p_mom = np.mean(ccnjSum) / np.max(ccnjSum)
n_mom = np.max(ccnjSum) / p_mom
moments_label = 'n = {0:.2f}\np = {1:.2f}'.format(n_mom, p_mom)

# fit binomial distribution using maximum likelihood estimation
p_mle = sum(ccnjSum) / (len(ccnjSum) * max(ccnjSum))
n_mle = max(ccnjSum) / p_mle
mle_label = 'n = {0:.2f}\np = {1:.2f}'.format(n_mle, p_mle)

x = np.arange(0, max(ccnjSum)+1)
plt.plot(x, binom.pmf(x, n_mom, p_mom), 'r-', label=moments_label)
plt.plot(x, binom.pmf(x, n_mle, p_mle), 'g--', label=mle_label)
plt.legend()
plt.show()
```
