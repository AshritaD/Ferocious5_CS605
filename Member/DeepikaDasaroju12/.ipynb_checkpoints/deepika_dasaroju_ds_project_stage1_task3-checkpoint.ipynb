In the data head we see the Delaware Kent County current_votes is 84975 where as the sum of total Kent County is around 40k. Let us verify the data for a single county

````python
def get_county_votes_diff_data(state, county):
  polled_votes = senate_merge_data[(senate_merge_data['state']==state) & (senate_merge_data['county']==county)]['current_votes'].item()
  candidate_total_votes = senate_county_candidate[(senate_county_candidate['state']==state) & (senate_county_candidate['county']==county)].groupby(by=['state','county'])['total_votes'].sum().item()
  return polled_votes-candidate_total_votes
````

````python
print(f"Delaware Kent County vote diff: {get_county_votes_diff_data('Delaware','Kent County')}")
print(f"Alabama Cullman County vote diff: {get_county_votes_diff_data('Alabama','Cullman County')}")
````

When verified this data online we see that the winner of the senate for Delaware Kent County got 43893 votes and Alabama Cullman County got 35949 votes. So this senate_county_candidate dataset has candidates who lost the elections. So We need to replace the winner votes by summing the loser's votes and getting the party name which is not listed. As in US most of the areas will have either Democratic Leader or Republic Leader. We will just check for those party names. So there can be minor errors in the data where both these party names appear and if the winner is an independent

````python
 def add_winner_and_votes(row):
  total_loser_votes = senate_county_candidate[(senate_county_candidate['state']==row['state'])&(senate_county_candidate['county']==row['county'])].groupby(by=['state','county'])['total_votes'].sum().item()
  loss_party_list = senate_county_candidate[(senate_county_candidate['state']==row['state'])&(senate_county_candidate['county']==row['county'])]['party'].to_list()
  if 'REP' not in loss_party_list:
    winner_party = 'REP'
  elif 'DEM' not in loss_party_list:
    winner_party = 'DEM'
  else:
    winner_party = 'OTHER'
  row['Winning_Party'] = winner_party
  row['Winner_Votes'] = row['current_votes'] - total_loser_votes
  return row
````

````python
senate_merge_data = senate_merge_data.apply(add_winner_and_votes,axis=1)
````

The House of Representatives is elected at Congressional District level all over the 50 states. Each congressional district consists of couple of counties. Some may have parts of counties or cities in them. So it is difficult to either divide the districts to counties or join counties data to form district. So House Election data will be analyzed at state level for the simplicity

After observing the above data, the district names either have the words **'s district number or at large ** after the state_name. So let us extract the state name from the data and sum the metrics of a state to get the state wide results

````python
house_merge_data = house_state.copy()
````

````python
house_merge_data.head()
````

````python
def get_state_name(row):
  if "’s" in row['district']:
    state_name = row['district'].split("’")[0]
  else:
    state_name = row['district'].split(" at")[0]
  row['district'] = state_name
  return row
````

Let us join the candidate data to this merge data through columns of Number of wins by a party in the state (We consider Democratic party, Republic Party and Independent Party) wins and their vote share in the state

````python
def get_each_party_win_count_vote_share(data):
  state_name = data['state']
  state_summary_df = house_merge_candidate[house_merge_candidate.district.str.contains(state_name)].groupby(
                          by='party').aggregate({"won":sum,"total_votes":sum}).reset_index()
  output = {}
  for label, row in state_summary_df.iterrows():
    if row['party'] == 'DEM':
      data['dem_win'] = row['won']
      data['dem_total_votes'] = row['total_votes']
    elif row['party'] == 'REP':
      data['rep_win'] = row['won']
      data['rep_total_votes'] = row['total_votes']
    else:
      data['oth_win'] = row['won']
      data['oth_total_votes'] = row['total_votes']
  return data
````
