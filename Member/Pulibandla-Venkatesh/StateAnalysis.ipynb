```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

```python
# Reading Covid cases confirmed data and displaying it

cases = pd.read_csv("../../DATASETS/COVID DATASETS/covid_confirmed_usafacts.csv")
cases 
```

```python
# Since I have to choose a particular state, i have grouped the data by state  name and made sum of all the cases in that state
# Since 'countyFIPS' and 'StateFIPS' are also interger type they also got added, which we will ignore.

state = cases.groupby(['State']).sum()
state
```

```python
# I have choose the state virginia.

virginia = state.loc['VA']
virginia # This is a series not a dataframe 
```

```python
# Plotting graph for last week of covid cases.

cases_graph = virginia.tail(7)
cases_graph.plot()

# The cases in the last week are stable.
```
