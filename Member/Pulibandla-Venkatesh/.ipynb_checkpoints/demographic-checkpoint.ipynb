```python
import pandas as pd
import numpy as np
```

```python
# Here i have displayed the enrichment dataset first 5 rows.

file_cvs = pd.read_csv("../../DATASETS/ENRICHMENT DATASETS/demographic updated.csv")
demographic = pd.DataFrame(file_cvs)
demographic.head()
```

```python
# In this step we have used the split() function to split the GEo_ID column so that i can get 'countyFIPS' column.
# 'countyFIPS' column will be helpfull to merge this dataset with covid-19 dataset.

demographic[['geoId','countyFIPS']] = demographic.GEO_ID.str.split("S", expand = True)
```

```python
# Here we have placed the 'countyFIPS' as a starting column using pop() and insert() functions.

countyID = demographic.pop('countyFIPS')
demograhic=demographic.insert(0, 'countyFIPS', countyID)
```

```python
demographic.dtypes
```

```python
# Here we have removed the extra column named 'geoId', which came into existence when we split the 'GEO_ID' column.

demographic.drop('geoId', axis=1,inplace = True)
```

```python
# here we have countyFIPS column to string type so that i can perform any string operations in future steps.

demographic.countyFIPS[0]= '00000'
demographic['countyFIPS']=demographic['countyFIPS'].astype(str)
```

```python
# Here we have filtered the data by removing the unwanted first row.

demographic.drop(0, inplace=True)
```

```python
# Here we have removed any extra zeros in the countyFIPS columns, so that we donot get any merge errors.

i=1
for id in demographic.countyFIPS:
    if id[0]=='0':
        y=id[1:]
        demographic.at[i,'countyFIPS'] = y
        i=i+1
        
demographic # This is the final dataset after filtering the data.
```

```python
# Reading the covid super dataset.

superCovidDS = pd.read_csv("../../Team/STAGE1/superCovidDS.csv")
```

```python
# Converting the 'countyFIPS' column to integer type to perform merge operations.

demographic['countyFIPS']=demographic['countyFIPS'].astype(int)
enrichment=pd.merge(superCovidDS,demographic, on =('countyFIPS'))
```
