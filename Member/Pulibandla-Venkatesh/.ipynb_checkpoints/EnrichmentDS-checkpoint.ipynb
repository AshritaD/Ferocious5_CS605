```python
# Importing numpy amd pandas packages

import numpy as np
import pandas as pd
```

```python
# Reading the enrichment data csv file and displaying it.

demographic = pd.read_csv("../../DATASETS/ENRICHMENT DATASETS/ACSDP1Y2021.DP05-Data.csv")
demographic
```

```python
# Removing the unwanted columns.
# Data filtering

for id in demographic.columns:
    if (id[-2:]=='EA' or id[-1]=='M' or id[-2:]=='MA' or id[-2:]=='PE' or id[-2:]=='PM'):
        del demographic[id]

demographic
    
```

```python
# removing the unwanted '0' row.

demographic.drop(0, inplace= True)
```

```python
demographic
```

```python
# data filtering and reading only wanted columns.


demographicNew = demographic.loc[:,('GEO_ID','NAME','DP05_0001E','DP05_0002E','DP05_0003E','DP05_0004E','DP05_0005E','DP05_0006E','DP05_0007E','DP05_0008E','DP05_0009E','DP05_0010E','DP05_0011E','DP05_0012E','DP05_0013E','DP05_0014E','DP05_0015E','DP05_0016E','DP05_0017E','DP05_0018E','DP05_0037E','DP05_0038E','DP05_0039E','DP05_0044E','DP05_0045E','DP05_0052E','DP05_0086E','DP05_0087E','DP05_0088E','DP05_0089E')]
```

```python
demographicNew
```

```python
# Names dictionary for the columns in a user readable way.

new_names={'DP05_0001E':'Total Population',
           'DP05_0002E':'Male population',
           'DP05_0003E':'Female Population',
           'DP05_0004E':'Sex Ratio',
           'DP05_0005E':'Population under 5 Years',
           'DP05_0006E':'Population 5 to 9 Years',
           'DP05_0007E':'Population 10 to 14 Years',
           'DP05_0008E':'Population 15 to 19 Years',
           'DP05_0009E':'Population 20 to 24 Years',
           'DP05_0010E':'Population 25 to 34 Years',
           'DP05_0011E':'Population 35 to 44 Years',
           'DP05_0012E':'Population 45 to 54 Years',
           'DP05_0013E':'Population 55 to 59 Years',
           'DP05_0014E':'Population 60 to 64 Years',
           'DP05_0015E':'Population 65 to 74 Years',
           'DP05_0016E':'Population 75 to 84 Years',
           'DP05_0017E':'Population Above 85 Years',
           'DP05_0018E':'Meadin Age Population',
           'DP05_0037E':'Population of White Race',
           'DP05_0038E':'Population of Black Race',
           'DP05_0039E':'Population of American Indian and Alaska Native',
           'DP05_0044E':'Population of Asian Race',
           'DP05_0045E':'Population of Indian Race',
           'DP05_0052E':'Population of Native Hawaiian and Other Pacific Islander',
           'DP05_0086E':'Total Housing Units',
           'DP05_0087E':'Population of Citizen Over 18',
           'DP05_0088E':'Population of Male Citizen',
           'DP05_0089E':'Population of Female Citizen'
          }
```

```python
# Reanaming the columns in the dataframe

demographicNew.rename(columns=new_names, inplace = True)
```

```python
demographicNew
```

```python
# Splitting the 'GEO_ID' column to get 'countyFIPS'

demographicNew[['geo_id','countyFIPS']] = demographicNew.GEO_ID.str.split('S',expand = True)
```

```python
# Placing the countyFIPS column at 0th position.

countyID = demographicNew.pop('countyFIPS')
demographicNew.insert(0,'countyFIPS',countyID)
```

```python
# Dropping the unwanted 'geo_id' column.

demographicNew.drop('geo_id',axis =1, inplace=True)
```

```python
# converting 'countyFIPS' to string type to perform operations.

demographicNew['countyFIPS'] = demographicNew['countyFIPS'].astype(str)
```

```python
# performing slicing operations to remove extra zero at the starting of 'countyFIPS' column.

i=1
for id in demographicNew.countyFIPS:
    if id[0]=='0':
        y = id[1:]
        demographicNew.at[i,'countyFIPS']=y
        i+=1
demographicNew
```

```python
# Reading the SuperCovid Dataframe

superCovidDS = pd.read_csv("../../Team/STAGE1/superCovidDS.csv")
```

```python
# Converting 'countyFIPS' in to integer type to perform merging.

demographicNew['countyFIPS']=demographicNew['countyFIPS'].astype(int)
```

```python
# Merging the supercovid data with demographic data.

enrichmentDemographic = pd.merge(superCovidDS,demographicNew, on ='countyFIPS')
```
