## Stage-1 (Task : 3) Performing intial merges with COVID-19 data using the variables in enrichment data

Description: In the task, Area code in Employment data set can be mapped to CountyFIPS in COVID-19 data set.

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Read the first Excel file 
df1 =  pd.read_csv(r"..\..\Desktop\DSA\superCovidDS.csv") 

# Read the second Excel file 
df2 = pd.read_excel(r'..\..\Desktop\DSA\allhlcn221.xlsx') 

filtered_df = df2[df2["Area Type"] == "County"]

filtered_df['Area\nCode']=filtered_df['Area\nCode'].astype(str).astype(int)

merged_df = pd.merge(filtered_df, df1, left_on="Area\nCode", right_on="countyFIPS", how="left")
merged_df.head(500)
```
